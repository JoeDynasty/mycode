```python
#tip calculator

print("INPUT")
print()

def cost_total():
    while True: 
        try:
            cost = float(input("Cost of meal: "))
            return cost
            if cost <= 0:
                print("Must be greater than 0. Please Try Again")
            else: 
                break
        except Exception as e:
            print("Invalid decimal number. Please try again.")    

def tip_total():
    while True:
        try:
            tip_percent = int(input("Tip percent: "))
            return tip_percent
        except Exception as e:
            print("Invalid integer. Please try again.")
            
def main(): 
    c = cost_total()
    t = tip_total()
    tip = c * (t / 100) #tip = cost * (tip_percent / 100)
    tip = round(tip, 2) #round to 2 places
    
    print()
    print("OUTPUT")
    print("Cost of meal: ", c)
    print("Tip percent: ", t , "%")
    print("Tip amount: " , tip)
    print("Total amount: " , c + tip)
    
if __name__ == "__main__":
    main()
```

```python
import random

ITEMS_FILENAME = "wizard_all_items.txt"
INVENTORY_FILENAME = "wizard_inventory.txt"

def read_items():
    try:
        items = []
        with open(ITEMS_FILENAME) as file:
            for line in file:
                line = line.replace("\n", "")
                items.append(line)
        return items
    except Exception as e:
        print("Could not find inventory file!")
        print("Exiting programs. Bye!")
        
        
def read_inventory():
    try:
        inventory = []
        with open(INVENTORY_FILENAME) as file:
            for line in file:
                line = line.replace("\n", "")
                inventory.append(line)
        return inventory
    except Exception as e:
        print("Could not find inventory file!")
        print("Wizzard is starting with no inventory")  

def write_inventory(inventory):
    try:
        with open(INVENTORY_FILENAME, "w") as file:
            for item in inventory:
                file.write(item + "\n")
    except Exception as e:
        print("Could not find inventory file!")
        print("Wizzard is starting with no inventory")  

def display_title():
    print("The Wizard Inventory program")
    print()    

def display_menu():
    print("COMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()

def walk(inventory):
    items = read_items()
    remaining_items = [i for i in items if i not in inventory]
    item = random.choice(remaining_items)
    print(f"While walking down a path, you see {item}.")
    choice = input("Do you want to grab it? (y/n): ")
    if choice == "y":
        if len(inventory) >= 4:
            print("You can't carry any more items. Drop something first.\n")
        else:
            inventory.append(item)
            print(f"You picked up {item}.\n")
            write_inventory(inventory)

def show(inventory):
    for number, item in enumerate(inventory, start=1):
        print(f"{number}. {item}")
    print()

def drop_item(inventory):
    while True:
        try:
            number = int(input("Number: "))
            if number < 1 or number > len(inventory):
                print("Invalid item number.\n")
            else:
                item = inventory.pop(number-1)
                print(f"You dropped {item}.\n")
                write_inventory(inventory)
                break
        except ValueError:
            print("Invalid item number. Please try again.")

def main():
    display_title()
    display_menu()

    inventory = read_inventory()
    while True:        
        command = input("Command: ")        
        if command == "walk":
            walk(inventory)
        elif command == "show":
            show(inventory)
        elif command == "drop":
            drop_item(inventory)
        elif command == "exit":
            break
        else:
            print("Not a valid command. Please try again.\n")
    print("Bye!")

if __name__ == "__main__":
    main()
```
